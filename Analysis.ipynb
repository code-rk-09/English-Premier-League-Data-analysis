```python
# Import all require libralies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('EPL_20_21.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Finding null values
df.isnull().sum()
```

```python
# Creating a two new column 
df['MinsPerMatch'] = (df['Mins']/df['Matches']).astype(int)
df['GoalsperMatch'] = (df['Goals']/df['Matches']).astype(float)
```

```python
# Succesfully add 2 new column
df.head()
```

```python
# Total Goal in legue
Total_Goals = df['Goals'].sum()
print(Total_Goals)
```

```python
# Total Penaltys
Total_Penalty_Goals = df['Penalty_Goals'].sum()
print(Total_Penalty_Goals)
```

```python
# Attempted Penaltys
Total_Penalty_Attempted = df['Penalty_Attempted'].sum()
print(Total_Penalty_Attempted)
```

```python
# Pie chart for penalties missed vs scored
plt.figure(figsize=(13,6))
penalty_not_scored = Total_Penalty_Attempted - Total_Penalty_Goals
data = [penalty_not_scored, Total_Penalty_Goals]
labels = ['Penalties missed', 'Penalties Scored']
color = sns.color_palette("Paired")
plt.pie(data,labels=labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
# Differnt positions
df['Position'].unique()
```

```python
# Finding how many player play as 'FW' Position
df[df['Position']=='FW']

```

```python
# Player coming from differnt countrys
np.size(df['Nationality'].unique())
```

```python
# Most player from which country
Nationality = df.groupby("Nationality").size().sort_values(ascending = False)
Nationality.head(10).plot(kind='bar',figsize=(12,6),color=sns.color_palette("rocket"))
```

```python
# Which is top 5 club has more number of players
df['Club'].value_counts().nlargest(5).plot(kind='bar',color=sns.color_palette("mako"))
```

```python
# Which is top 5 club has least number of players
df['Club'].value_counts().nsmallest(5).plot(kind='bar',color=sns.color_palette("hls"))
```

```python
# Player based on Age Groups
Under20 = df[df['Age'] <=20]
Age_20_to_25 = df[(df['Age']>20) & (df['Age']<=25)]
Age_25_to_30 = df[(df['Age']>25) & (df['Age']<=30)]
Above30 = df[df['Age']>30]
```

```python
x = np.array([Under20['Name'].count(),Age_20_to_25['Name'].count(),Age_25_to_30['Name'].count(),Above30['Name'].count()])
mylabels = ["Under20","20-25","25-30","Above30"]
plt.title('Total Player with Age',fontsize = 20)
plt.pie(x, labels = mylabels, autopct="%.1f%%")
plt.show()
```

```python
# Total player under age of 20 in each club
Players_under_20 = df[df['Age']<20]
Players_under_20['Club'].value_counts().plot(kind='bar',color=sns.color_palette("Paired"))
```

```python
# List of player under age of 20 in Manchester United 
Players_under_20[Players_under_20["Club"]=='Manchester United']
```

```python
# Average age of player in each group
Num_player = df.groupby("Club").size()
data = (df.groupby('Club')['Age'].sum())/Num_player
data.sort_values(ascending=False)
```

```python
# Visuallizing avarage age group
plt.figure(figsize=(12,6))
sns.boxplot(x='Club', y='Age', data=df)
plt.xticks(rotation = 90)
```

```python
# Total assistance from each club
Assists_by_clubs = pd.DataFrame(df.groupby('Club',as_index=False)['Assists'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax = sns.barplot(x='Club',y='Assists',data=Assists_by_clubs.sort_values(by="Assists"),palette="icefire")
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation=90)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of clubs vs Total Assists', fontsize=20)
```

```python
# Top 10 Assists
top_10_assists = df[['Name', 'Club', 'Assists', 'Matches']].nlargest (n=10, columns= 'Assists') 
top_10_assists
```

```python
# Total goals from each club
Goals_by_clubs = pd.DataFrame(df.groupby('Club', as_index=False)['Goals'].sum()) 
sns.set_theme(style="whitegrid",color_codes=True) 
ax=sns.barplot(x='Club',y='Goals', data=Goals_by_clubs.sort_values (by="Goals"), palette="Spectral") 
ax.set_xlabel("club", fontsize=35) 
ax.set_ylabel("Goals", fontsize=20)
plt.xticks(rotation=90) 
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs vs Total Goals', fontsize = 20)
```

```python
# Most goal by player
top_10_Goals = df[['Name', 'Club', 'Goals', 'Matches']].nlargest (n=10, columns= 'Goals') 
top_10_Goals
```

```python
# Goals per match
top_10_Goals_Per_Match = df[['Name', 'GoalsperMatch', 'Goals', 'Matches']].nlargest (n=10, columns= 'GoalsperMatch') 
top_10_Goals_Per_Match
```

```python
# Pie chart of Goals with assist and without assist
plt.figure(figsize=(14,7))

assists = df['Assists'].sum()

data= [Total_Goals - assists, assists]

labels = ['Goals without assists', 'Goals with assists'] 
color = sns.color_palette("Paired")

plt.pie(data, labels = labels, colors = color, autopct = "%.0f%%") 
plt.show()
```
